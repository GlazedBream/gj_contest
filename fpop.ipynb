```python
# 기본 제공 라이브러리 pandas, numpy, openpyxl, pyarrow, geopandas, matplotlib, tqdm
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# 실제 환경에서 유동인구: r"C:\Users\User\anaconda_projects\data\fpop_2023MM.csv", 카드매출: r"C:\Users\User\anaconda_projects\data\card_2023MM.csv"
df = pd.read_csv("data/2324_유동인구_샘플/202412_유동인구_샘플.csv")
print(df.shape)

df
```

```python
shp_path = "data/shp/100m격자.shp"
gdf = gpd.read_file(shp_path)
gdf.rename(columns={"gid": "GID"}, inplace=True)
gdf
```

```python
shp_path3 = r"data\shp\N3A_G0110000.shp"
gdf3 = gpd.read_file(shp_path3)
gdf3
```

```python
gdf_fpop = gdf.merge(df, on="GID", how="right")
print(gdf_fpop.shape)
gdf_fpop.head()
```

```python
fig, ax = plt.subplots(figsize=(6, 6), facecolor="white")
ax.axis("off")
gdf_fpop.boundary.plot(ax=ax, color="white", linewidth=0.5)

# 동 경계
gdf3.boundary.plot(ax=ax, color="black", linewidth=0.5)

# 관심 컬럼을 지정
gdf_fpop.plot(column="TOTAL", cmap="coolwarm", ax=ax)
plt.tight_layout()
# plt.savefig("test.png", dpi=600, transparent=True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6), facecolor="white")
ax.axis("off")
gdf_fpop.boundary.plot(ax=ax, color="white", linewidth=0.5)

# 동 경계
gdf3.boundary.plot(ax=ax, color="black", linewidth=0.5)

# 관심 컬럼을 지정
gdf_fpop.plot(column="TOTAL", cmap="coolwarm", ax=ax)
plt.tight_layout()
# plt.savefig("test.png", dpi=600, transparent=True)
plt.show()
```

```python
# Tukey

q1 = gdf_fpop["TOTAL"].quantile(0.25)
q3 = gdf_fpop["TOTAL"].quantile(0.75)
iqr = q3 - q1

upper_bound = q3 + 1.5 * iqr
lower_bound = q1 - 1.5 * iqr

gdf_fpop2 = gdf_fpop[
    (gdf_fpop["TOTAL"] >= lower_bound) & (gdf_fpop["TOTAL"] <= upper_bound)
]
print(gdf_fpop.shape)
print(gdf_fpop2.shape)
```

```python
# Tukey 이상치 제거 후 한 번 더 그리기

fig, ax = plt.subplots(figsize=(6, 6), facecolor="white")
ax.axis("off")
gdf_fpop2.boundary.plot(ax=ax, color="white", linewidth=0.5)

# 동 경계
gdf3.boundary.plot(ax=ax, color="black", linewidth=0.5)

# 관심 컬럼을 지정
gdf_fpop2.plot(column="TOTAL", cmap="coolwarm", ax=ax)
plt.tight_layout()
# plt.savefig("test.png", dpi=600, transparent=True)
plt.show()
```
